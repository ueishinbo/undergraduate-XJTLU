```python
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']

datas = pd.read_csv('Data.csv', encoding='utf-8')

long_time = list(datas['时长'])

d = 10
num_bins = int((max(long_time) - min(long_time)) / d)

plt.hist(long_time, range(min(long_time), max(long_time) + d, d), density=True, color="#33ff44")

plt.xticks(range(min(long_time), max(long_time) + d, d))
plt.grid(alpha=0.4)

plt.title('The Movie Duration Distribution')
plt.xlabel('Duration')
plt.ylabel('Proportion')
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif']=['SimHei']
datas = pd.read_csv('Data.csv', encoding='UTF-8')

datas['country'] = datas['国家']
# print(datas['country'])

country = datas.groupby('country')['country'].count()
# print(country)
# print(list(country))
# print(list(country.index))
explods = [0, 0.2, 0, 0, 0, 0, 0, 0, 0]

plt.pie(list(country), labels=list(country.index), autopct='%1.1f%%')
plt.title('电影的国家/地区分布情况')
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter

plt.rcParams['font.sans-serif']=['SimHei']

datas = pd.read_csv('Data.csv', encoding='utf-8')
s = ''
for i in range(99):
    s += datas.iloc[i, 2]+','
s += datas.iloc[99, 1]
# print(s)
authors = s.split(',')
# print(authors)
c = Counter(authors)
# print(c)
items = c.most_common(6)
# print(items)

author = []
count = []
for item in items:
    author.append(item[0])
    count.append(item[1])

# print(author)
# print(count)
plt.bar(author, count, color=['#33ff00', '#33ff33', '#33ff66', '#33ff99', '#33ffcc', '#33ffff'])
plt.title('Occurrences Order For Stars')
plt.xlabel('Actor')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']

datas = pd.read_csv('Data.csv', encoding='utf-8')

datas['year'] = datas['上映时间']


year = datas.groupby('year')['year'].count()
plt.figure(figsize=(20, 10), dpi=80)

plt.plot(list(year.index), list(year))
plt.title('电影年份的分布情况')
plt.xlabel('年份')
plt.ylabel('电影数量')
plt.grid(alpha=0.9)
plt.show()
```
